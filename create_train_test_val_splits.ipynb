```python
for i in range(1, 20000):
    path = random.choice(os.listdir('images_128x128/'))
    if i < 2000:
        os.rename("images_128x128/"+path, "val_images_128x128/val_"+str(i).zfill(5)+".png")
        os.rename("images_64x64/"+path, "val_images_64x64/val_"+str(i).zfill(5)+".png")
    elif i < 4000:
        os.rename("images_128x128/"+path, "test_images_128x128/val_"+str(i).zfill(5)+".png")
        os.rename("images_64x64/"+path, "test_images_64x64/val_"+str(i).zfill(5)+".png")
    else:
        os.rename("images_128x128/"+path, "train_images_128x128/val_"+str(i).zfill(5)+".png")
        os.rename("images_64x64/"+path, "train_images_64x64/val_"+str(i).zfill(5)+".png")
```
